### *Objective y Purpose*  
*Objective*: To qualify the neigborhoods in Mexico City based on the availability of services or venues.  
*Purpose*: The identification of the best neighborhoods to live in Mexico City.

### *Background*  
Mexico City is one of the largest metropolitan zones in the world. In 1900 the city size was less than 10% of what it's today and the growth in the following four decades ended by swallowing small towns that have became important turistry centers (like Coyoacan, or Tlalpan) or business spots (like Condesa and Santa Fe).  
Between 1950 and 1970 many people begin to move to new neighborhoods in the the city edges reaching a point to exceed the *Federal District* limits into some *State of Mexico* muncipalities, creating tranports problems because of the transfer form residential zones to workplaces.  
The city continue to grow as the same as service providers seek the best spots for their venues, creating differents zones of concentration, as well as more residential zones were created each time more far away from these zones, magnifying the transfer times up to 2-3 hours, in the morning as well as at the afertnoon.

The city is the country capital and by mid XIX century  an area from the State of Mexico was delimitated and called *Federal District (DF)* and was divided in 16 delegations (the equivalent to municipalities, boroughs or countys).   
When the urban stain exceeded the Federal District limits, the city was denominated as *Metropolitan Zone of Mexico City (ZMCM)* or *of Mexican Valley (ZMVM)*. In 2017 their distrcit status change to became a state and it was called state of *Mexico City (CDMX)* as well as the change of name from delegations to *town halls*.

### *Antecedentes*  
La Ciudad de México es una las zonas metropolitanas más grandes del mundo. En 1900 el tamaño de la ciudad era menor al 10% de su tamaño actual y el crecimiento de las siguientes cuatro décadas terminó por abarcar pequeños pueblos que se han convertido en importantes centros turísticos (como Coyoacán o Tlalpan) o centro de negocio (como La Condesa o Santa Fe).  
Entre 1950 y 1970 mucha gente comenzó a mudarse a nuevas colonias en las orillas de la ciudad de aquel entonces al grado de sobrepasar los límites del *Distrito Federal* abarcando algunos municipios del *Estado de Mexico* y creando problemas de transporte por el traslado desde las zonas  residenciales hasta las zonas de trabajo.  
La ciudad siguió creciendo al tiempo que los proveedores de servicio buscaron los mejor puntos para sus locales creando diferentes zonas de concentración al tiempo que se creaban zonas de residencia cada vez más alejados de estos zonas, magnificando los tiempos traslado hasta de 2-3 horas, tanto en las mañanas como en las tardes.
##### *Denominaciones y Aclaraciones*  
Cuando se habla de la *Ciudad de México* es fácil confundirse con CDMX, Distrito Federal, Zona Metropolitana del Valle de México, Estado de México y México. La ciudad es la capital del país y a medidado del siglo XIX fue delimitada una zona del Estado de México al que pertenecía llamádolo *Distrito Federal (DF)* y se dividió en 16 delegaciones (equivalentes a municipios en el resto del país).  
Al rebasar la mancha urbana los límites del Distrito Federal se denominó a la ciudad como *Zona Metropolitana de la Ciudad de México(ZMCM)* o *del Valle de México (ZMVM)*. En 2017 se cambió el estatus de distrito para convertirse en estado y se le denominó *Ciudad de México (CDMX)* al estado al tiempo de cambiar las delegaciones por *alcaldías*.  
En este trabajo no haremos el análisis del objetivo en base a la ZMVM que incluiría 16 alcaldías de la CDMX, 23 municipios del Estado de México y 1 muncipio de Hidalgo. Para este trabajo solo utilizaremos las información de las 16 alcaldías y nos referiremos a la urbe como *Ciudad de México* para no crear confusión con la ZMVM.

### *Problema*  
La ciudad, al ser una urbe con gran atractivo para establecerse, genera la pregunta típica de: ¿Dónde es el mejor lugar para vivir en la Ciudad de México? Para responder la pregunta debemos establecer primero cómo calificar a las colonias en base a sus servicios y darle valor a estos en base a los requerimientos de la personas interesada en identificar la mejor opción para vivir.  
Los parámetros no son iguales para todos, ya que los intereses pueden ser muy variados dependiendo de las condiciones de vida que requiren, nivel socioeconómico, edad, etc. Por ejemplo, una persona de edad mayor puede estar buscando una colonia que tenga parques, teatros y cafeterías, otra persona con hijos puede estar buscando colonias con escuelas, cines y centros comerciales, y otra persona puede estar buscando colonias que tengan establecimientos con música, bares y gimnasios. La combinación de intereses pueden ser tan variados como personas que viven en la ciudad.  
El projecto busca lograr el objetivo planetado para una combinación de intereses específicos (tomando el ejemplo de una persona) y que pueda adaptarse a cualquier otra integreando sus intereses particulares.

### *Interesados*  
Los interesados en este projecto pueden ser tanto personas que quieren establecerse en la Ciudad de México como aquellas que buscan mudarse a otras colonias por diferentes motivos, especialmente por el tiempo de traslado de sus residencias hasta sus zonas de trabajo. También puede servir para aquellos comercios que pueden identificar la plusvalía de sus servicios en determinadas colonias que carecen de éstos.

*Foursquare — ‘Places API’*: We will be use Foursquare’s API Places to acquire data related to ‘venues’ (as defined by Foursquare) as the 'interest's' of the particular person of example. Each Foursquare ‘venue’ is assigned a ‘category’ and each ‘category’ is associated with a particular ‘categoryID’.
